```python
from datasets import load_dataset
import re


dataset = load_dataset("ai4bharat/IndicQA", "indicqa.hi")

def combine_fields(example):
    context = example.get('context', '')
    question = example.get('question', '')
    answer = example.get('answers', {}).get('text', [''])[0]
    combined_text = f"Context: {context} Question: {question} Answer: {answer}"
    return {'combined_text': combined_text}

combined_dataset = None

for split in dataset.keys():
    combined_split = dataset[split].map(combine_fields, remove_columns=['context', 'question', 'answers'])
    if combined_dataset is None:
        combined_dataset = combined_split
    else:
        combined_dataset = combined_dataset.concatenate(combined_split)

text = " ".join(combined_dataset['combined_text'])
text = re.sub(r'\s+', ' ', text)

tokens = text.encode("utf-8")
tokens = list(map(int, tokens))

print(tokens[:500])
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
    newids = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i + 1] == pair[1]:
            newids.append(idx)
            i += 2
        else:
            newids.append(ids[i])
            i += 1
    return newids

# vocabulary size
vocab_size = 32000
num_merges = vocab_size - 256
ids = list(tokens)

merges = {}
for i in range(num_merges):
    stats = get_stats(ids)
    pair = max(stats, key=stats.get)
    idx = 256 + i
    print(f"merging {pair} into a new token {idx}")
    ids = merge(ids, pair, idx)
    merges[pair] = idx

print("tokens length:", len(tokens))
print("ids length:", len(ids))
print(f"compression ratio: {len(tokens)/len(ids):.5f}X")

vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]
```

## Creating vocab file
####The file will contain a dictionary where keys are token IDs (as strings) and values are lists of integers representing byte sequences--> To append to LLaMA's vocabulary, mapping of byte sequences back to strings

####LLaMA's tokenizer requires mapping of tokens to their corresponding string representations, not just byte sequences

```python
import json

vocab_file = "bpe_vocab.json"
with open(vocab_file, "w") as file:

    json.dump({str(k): v.decode("utf-8") for k, v in vocab.items()}, file)

print(f"Vocabulary saved to {vocab_file}")

```

##Append Llama vocab

```python
from transformers import LlamaTokenizer, LlamaForCausalLM
from transformers import AutoTokenizer, AutoModelForCausalLM
import json

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3.1-8B")

with open("bpe_vocab.json", "r") as file:
    new_vocab = json.load(file)

reverse_vocab = {v: k for k, v in new_vocab.items()}

tokenizer.add_tokens(list(reverse_vocab.keys()))

updated_tokenizer_file = "updated_tokenizer"
tokenizer.save_pretrained(updated_tokenizer_file)

model = LlamaForCausalLM.from_pretrained('meta-llama/Meta-Llama-3.1-8B-Instruct')

model.resize_token_embeddings(len(tokenizer))


updated_model_file = "updated_model"
model.save_pretrained(updated_model_file)

print(f"Updated tokenizer saved to {updated_tokenizer_file}")
print(f"Updated model saved to {updated_model_file}")
```

```python
def decode(ids):
    tokens = b"".join(vocab[idx] for idx in ids)
    text = tokens.decode("utf-8", errors="replace")
    return text

def encode(text):
    tokens = list(text.encode("utf-8"))
    while len(tokens) >= 2:
        stats = get_stats(tokens)
        pair = min(stats, key=lambda p: merges.get(p, float("inf")))
        if pair not in merges:
            break
        idx = merges[pair]
        tokens = merge(tokens, pair, idx)
    return tokens

# Example
print(encode("अंतरराष्ट्रीय व्यापार में प्रतिस्पर्धात्मकता बनाए रखने के लिए नवाचार और तकनीकी उन्नयन अनिवार्य हैं।"))
print(decode(encode("अंतरराष्ट्रीय व्यापार में प्रतिस्पर्धात्मकता बनाए रखने के लिए नवाचार और तकनीकी उन्नयन अनिवार्य हैं।")))
```
