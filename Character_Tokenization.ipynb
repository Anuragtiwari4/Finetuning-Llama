#Character Tokenization

```python

!pip install datasets

from datasets import load_dataset
import huggingface_hub

# Hugging Face token
HUGGINGFACE_TOKEN = "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"

huggingface_hub.login(HUGGINGFACE_TOKEN)

dataset = load_dataset("ai4bharat/IndicQA", "indicqa.hi")

print("Dataset Keys:", dataset.keys())
print("Example Data:", dataset[list(dataset.keys())[0]][0])

def char_tokenize(sentence):
    return list(sentence)

def tokenize_dataset(dataset, field='context'):
    tokenized_dataset = dataset.map(lambda examples: {'char_tokens': [char_tokenize(sentence) for sentence in examples[field]]}, batched=True)
    return tokenized_dataset

dataset_key = list(dataset.keys())[0]

tokenized_dataset = tokenize_dataset(dataset[dataset_key])

print("Original Sentence:", dataset[dataset_key][1]['context'])
print("Character Tokens:", tokenized_dataset[1]['char_tokens'])

# Save tokenized dataset
tokenized_dataset.save_to_disk(f'tokenized_hindi_{dataset_key}')
```

# SUBWORD TOKENIZATION

```python
!pip install datasets transformers

from datasets import load_dataset
from transformers import BertTokenizer
import huggingface_hub

HUGGINGFACE_TOKEN = "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"

huggingface_hub.login(HUGGINGFACE_TOKEN)

dataset = load_dataset("ai4bharat/IndicQA", "indicqa.hi")

tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

def subword_tokenize(sentence):
    return tokenizer.tokenize(sentence)

def tokenize_dataset(dataset, field='context'):
    tokenized_dataset = dataset.map(lambda examples: {'subword_tokens': [subword_tokenize(sentence) for sentence in examples[field]]}, batched=True)
    return tokenized_dataset

dataset_key = list(dataset.keys())[0]

tokenized_dataset = tokenize_dataset(dataset[dataset_key])

print("Original Sentence:", dataset[dataset_key][1]['context'])
print("Subword Tokens:", tokenized_dataset[1]['subword_tokens'])

tokenized_dataset.save_to_disk(f'tokenized_hindi_{dataset_key}_subword')
```

# WORD TOKENIZATION

```python
!pip install datasets

from datasets import load_dataset
import huggingface_hub

HUGGINGFACE_TOKEN = "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"

huggingface_hub.login(HUGGINGFACE_TOKEN)

dataset = load_dataset("ai4bharat/IndicQA", "indicqa.hi")

def word_tokenize(sentence):
    return sentence.split()

def tokenize_dataset(dataset, field='context'):
    tokenized_dataset = dataset.map(lambda examples: {'word_tokens': [word_tokenize(sentence) for sentence in examples[field]]}, batched=True)
    return tokenized_dataset

dataset_key = list(dataset.keys())[0]

tokenized_dataset = tokenize_dataset(dataset[dataset_key])

print("Original Sentence:", dataset[dataset_key][1]['context'])
print("Word Tokens:", tokenized_dataset[1]['word_tokens'])

tokenized_dataset.save_to_disk(f'tokenized_hindi_{dataset_key}_word')
```
